```python
import json
import pandas as pd

# Leer settings.json
with open("settings.json", "r") as f:
    settings_data = json.load(f)

settings_df = pd.DataFrame([settings_data])
print("Settings DataFrame:")
print(settings_df)

# Leer stats.json
with open("stats.json", "r") as f:
    stats_data = json.load(f)

stats_df = pd.DataFrame([stats_data])
print("\nStats DataFrame:")
print(stats_df)

# Leer metadata.json
with open("metadata.json", "r") as f:
    metadata_data = json.load(f)

metadata_df = pd.DataFrame([metadata_data])
print("\nMetadata DataFrame:")
print(metadata_df)
```

```python
# Para stocks
stocks_df = pd.DataFrame(stats_df["stocks"].iloc[0])

# Para conversions
conversions_df = pd.DataFrame(stats_df["conversions"].iloc[0])
conversions_df['conversion_id'] = range(1, len(conversions_df) + 1)
moves_list = []
for idx, row in conversions_df.iterrows():
    for move in row['moves']:
        move['conversion_id'] = row['conversion_id']
        moves_list.append(move)
moves_df = pd.DataFrame(moves_list)

# Para combos
combos_df = pd.DataFrame(stats_df["combos"].iloc[0])
combos_df['combo_id'] = range(1, len(combos_df) + 1)
combo_moves_list = []
for idx, row in combos_df.iterrows():
    for move in row['moves']:
        move['combo_id'] = row['combo_id']
        combo_moves_list.append(move)
combo_moves_df = pd.DataFrame(combo_moves_list)

# Para actionCounts
action_counts_df = pd.DataFrame(stats_df["actionCounts"].iloc[0])
columns_to_expand = ["lCancelCount", "attackCount", "grabCount", "throwCount", "groundTechCount", "wallTechCount"]
for col in columns_to_expand:
    expanded_data = action_counts_df[col].apply(pd.Series)
    expanded_data = expanded_data.add_prefix(f"{col}_")
    action_counts_df = pd.concat([action_counts_df.drop(col, axis=1), expanded_data], axis=1)

# Para overall
overall_df = pd.DataFrame(stats_df["overall"].iloc[0])
columns_to_expand_overall = ["inputCounts", "successfulConversions", "inputsPerMinute", 
                             "digitalInputsPerMinute", "openingsPerKill", "damagePerOpening", 
                             "neutralWinRatio", "counterHitRatio", "beneficialTradeRatio"]
for col in columns_to_expand_overall:
    expanded_data_overall = overall_df[col].apply(pd.Series)
    expanded_data_overall = expanded_data_overall.add_prefix(f"{col}_")
    overall_df = pd.concat([overall_df.drop(col, axis=1), expanded_data_overall], axis=1)
```
